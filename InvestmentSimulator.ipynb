```python

# Calculating returns
Ys = data_stock[assets_stock].pct_change().dropna()


Ys *= 100

stocks_to_research = {}

from tqdm import tqdm

# ticker = Ticker(["^IRX"])
# T = ticker.history("1mo")
# plt.plot(AAPL['dividends'].values)

for i in tqdm(range(50)):
    
    low = np.random.randint(0,252 * 8)
    high = low + 252 * 5
    
    
    Ysub = Ys.iloc[low:high,:]
    
    mus = Ysub.mean().to_numpy().reshape(1,-1)[0]


    sigmas = Ysub.cov().to_numpy()

    # sigma = np.append(sigma, np.zeros((sigma.shape[0], 1)), axis=1)
    # sigma = np.append(sigma, np.zeros((1, sigma.shape[1])), axis=0)


    rate_vals_stk = Ysub.values
    # assets.append("Short Term Treasury")

    #compute downside (semi)skewness:

    for i in range(len(mus)):
        val = 0
        cnt = 0
        for j in range(Ysub.shape[0]):
            if rate_vals_stk[j, i] < 0:
                val += ((rate_vals_stk[j,i] - mus[i])**4)
                cnt += 1
        val /= (cnt+1)


        sigmas[i, i] += .0001*val


    # Defining initial variables
    xs = cp.Variable((len(mus), 1))

    # Budget and weights constraints
    constraints_stk = [(252) * mus@xs >=35,cp.sum(xs) == 1,
                   xs <= 1,
                   xs >= 0]

    # Defining risk objective
    risks = cp.quad_form(xs, sigmas)
    objectives = cp.Minimize(risks)

    weights = pd.DataFrame([])
    # Solving the problem with several solvers
    probs = cp.Problem(objectives, constraints_stk)
    probs.solve()


    weights_stock = list(zip(assets_stock,xs.value))

    weights_stock = sorted(weights_stock, key = lambda y: y[1][0], reverse = True)

    for w in weights_stock:
        
        if w[1][0] > .01:
            if w[0] in stocks_to_research:
                stocks_to_research[w[0]] +=1
            else:
                stocks_to_research[w[0]] =1
                
# print(stocks_to_research)

sorted_stocks = sorted(stocks_to_research.items(), key = lambda x: x[1], reverse = True)

for s in sorted_stocks:
    print(s[0], s[1])

#     print(np.sum(xs.value.T[0]))


#     print(252 * xs.value.T[0].dot(mus))
    

```
